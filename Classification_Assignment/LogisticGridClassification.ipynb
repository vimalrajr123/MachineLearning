```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
independent=dataset.drop(columns=['classification'])
dependent=dataset['classification']
```

```python
independent = pd.get_dummies(independent, drop_first=True)
```

```python
independent
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
# Define the parameter grid
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100], # Regularization strength
    'penalty': ['l1', 'l2', 'elasticnet', 'none'], # Penalty type
    'solver': ['lbfgs', 'liblinear', 'saga'], # Solvers compatible with penalties
}
```

```python
from sklearn.linear_model import LogisticRegression
grid = GridSearchCV(LogisticRegression(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train, y_train)
re=grid.cv_results_
#print(re)
```

```python
print("Best Parameters:", grid.best_params_)
print("Best Cross Validation Score:", grid.best_score_)
```

```python
grid_predictions = grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions,labels=['yes', 'no'])
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')
```

```python
print("F1 macro:", f1_macro)
print("The confusion Matrix:\n",cm)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc=roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])
print("ROC_AUC Score:", roc_auc)
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
# print classification report
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
print(clf_report)
```

```python
print("OVERALL REPORT for Logistic Regression Classification:\n")
print("Best Parameters:\n\t", grid.best_params_)
print("Best Cross Validation Score:\n\t", grid.best_score_)
print("F1 macro:\n\t", f1_macro)
print("The confusion Matrix:\n",cm)
print("ROC_AUC Score:\n\t", roc_auc)
print("classification_report:\n\t",clf_report)
```
