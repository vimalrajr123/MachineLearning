```python
# pandas is a library in python used for handling and preparing the data before training the model

import pandas as pd   # here we are creating alias for pandas library which is "pd"
```

```python
dataset = pd.read_csv("Salary_Data.csv")  #read_csv is a function in pandas used for reading the csv data and return 
                                            # as dataframe and we are storing it in the dataset variable
```

```python
dataset #print dataset
```

```python
# now split the input and output assign to independant and dependant variables from the dataset 
# here YearsExperience is the input and Salary is the output
independant = dataset[["YearsExperience"]]  
dependant = dataset[["Salary"]]
```

```python
independant
```

```python
dependant
```

```python
# now spilt train dataset and test dataset for this we use sklearn library which has most of the machine learing module, class and functions
# sklear is the library, model_selection is the module 
# here train_test_split splits the dataset into train and test set and returns in the form of  X_train, X_test, y_train, y_test
#X_train = input data for training, X_test = input data for testing, y_train= output data for training, y_test = output data for testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independant,dependant,test_size=0.30,random_state=0)
```

```python
#LinearRegression is a class create a object "regressor"
#linear_model is a module that has model creations algorithms
#though our output it continuous number / numeric it is regression and we have only one input so its simple Linear Regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
#fit function helps model to learn pattern from input data and correct answers in our case (X_train and y_train)
# and give the best values solpe(m) and intercept(b)
regressor.fit(X_train,y_train)
```

```python
#we can acess the bestvalues like below in our case coeficient and intercept
weight = regressor.coef_
weight
```

```python
bias = regressor.intercept_
bias
```

```python
#now our model has the constants already use it to predit using the testset
y_pred = regressor.predict(X_test) 
```

```python
#now need to check if our model is good model or bad one
#for evaluation metrics, the validation parameter we are using is r2 , pass the output testset and predicted output
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#if r2 score is near 1 then its a good model if its near 0 then its a bad model
#in our case its near 1 so its good model.
#so we gonna save the model and load it for later use
```

```python
#save the model into sav file
#pickle is a inbuilt python module, pickel save the python object into a file
import pickle
filename = "finalized_model_linear.sav" 
pickle.dump(regressor,open(filename,'wb')) #writing the python object regressor into a file dump() used to write object into binary format into the file
            
```
